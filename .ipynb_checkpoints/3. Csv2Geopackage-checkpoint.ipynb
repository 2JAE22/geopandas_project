```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import os
```

```python
path = "data"
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path,'대구광역시_시내버스 정류소 위치정보_20220731.csv'), encoding = "cp949")
df
```

```python
df.columns
```

```python
df = df.drop(['정류소아이디', '모바일아이디',  '영문명', '시도', '구군', '동', '경유노선',],axis = 1)
df
```

```python
geometry = gpd.points_from_xy(df["경도"],df["위도"])
gdf = gpd.GeoDataFrame(df, geometry = geometry)
gdf
```

```python
ax = gdf.plot(figsize = (10,8))
ax.set_title("Bus stop")
#ax.set_axis_off()
plt.show()
```

```python
gdf.crs #Coordinate Refernce System_좌표참조시스템
```

```python
gdf = gdf.set_crs(epsg = 4326) #european Petroleum Survey Group_각 CRS 를 고유하게 식별하는 번호임
gdf.crs
```

```python
gdf.to_file(os.path.join(path, "대구시_버스정류장.gpkg"),drive = "GPKG", encoding = "UTF-8")
```
