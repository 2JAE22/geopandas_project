```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import os
```

```python
path = "data"
os.listdir(path)
```

```python
gdf = gpd.read_file(os.path.join(path, '대구시_시군구.gpkg'))
gdf = gdf.to_crs(epsg=5179)
gdf.head()
```

```python
gdf.total_bounds #x최솟값 x최댓값 y최솟값 y최댓값
```

```python
x_min,x_max,y_min,y_max = gdf.total_bounds
```

```python
width = 1000
length=1000
cols = list(range(int(np.floor(x_min)), int(np.floor(x_max)),width))
rows = list(range(int(np.floor(y_min)), int(np.floor(y_max)),length))

rows.reverse()
plygons = []

for x in cols:
    for y in rows:
        temp = Polygon([
        (x,y),
        (x+width,y),
        (x+width,y-length),
        (x,y-length),]
        )
        plygons.append(temp)
    
plygons
```

```python
grid = gpd.GeoDataFrame({"geometry":plygons})
grid.crs = "epsg:5179"
grid["id"] = grid.index
grid
```
